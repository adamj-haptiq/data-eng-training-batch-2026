```python
# 10. How to keep only top 2 most frequent values as it is and replace everything else as ‘Other’?

import numpy as np
import pandas as pd
np.random.seed(100)
ser = pd.Series(np.random.randint(1, 5, [12]))

top2 = ser.value_counts().nlargest(2).index
print(ser.where(ser.isin(top2), "Other"))
```

```python
# 22. How to get the day of month, week number, day of year and day of week from a series of date strings?

import numpy as np
import pandas as pd

ser = pd.Series(
    [
        "01 Jan 2010",
        "02-02-2011",
        "20120303",
        "2013/04/04",
        "2014-05-05",
        "2015-06-06T12:20",
    ]
)

dates = pd.to_datetime(ser, format="mixed")
result = pd.DataFrame(
    {
        "day": dates.dt.day,
        "iso_week": dates.dt.isocalendar()["week"],
        "day_of_year": dates.dt.dayofyear,
        "day_of_week": dates.dt.dayofweek,  # Monday=0 ... Sunday=6
    }
)
print(result)
```

```python
# 25. How to filter valid emails from a series?

import pandas as pd

ser = pd.Series(
    [
        "user@example.com",
        "plain text",
        "no-at-sign.com",
        "@nodomain.com",
        "ok+tag@mail.co.uk",
        "bad@@double.com",
        "name_99@sub.domain.org",
    ]
)

# Practical pattern: local@domain.tld (not a full RFC parser)
email_re = r"^[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,}$"
valid = ser[ser.str.fullmatch(email_re, na=False)]
print(valid)
```

```python
# 51. How to get the row number of the nth largest value in a column?

import numpy as np
import pandas as pd

df = pd.DataFrame(np.random.randint(1, 30, 30).reshape(10, -1), columns=list("abc"))

n = 5
col = "a"

# Row index *label* of the n-th largest value in `col` (stable tie-breaking like nlargest)
nth_label = df[col].nlargest(n).index[-1]

# Integer position of that row in `df` (useful if the index is not 0..N-1)
nth_row = df.index.get_loc(nth_label)

print("nth largest value:", df.loc[nth_label, col])
print("row index label:", nth_label)
print("row number (position):", nth_row)
```

```python
# 74. How to get the frequency of unique values in the entire dataframe?

import numpy as np
import pandas as pd

df = pd.DataFrame(np.random.randint(1, 10, 20).reshape(-1, 4), columns=list("abcd"))

# Count each distinct value across every cell (order of columns/rows does not matter)
freq = df.stack().value_counts()
print(freq)

# Equivalent: pd.Series(df.to_numpy().ravel()).value_counts()
```

```python
# 75. How to split a text column into two separate columns?
import pandas as pd
df = pd.DataFrame(["STD, City State",
"33, Kolkata West Bengal",
"44, Chennai Tamil Nadu",
"40, Hyderabad Telengana",
"80, Bangalore Karnataka"], columns=['row'])

print(df)
#> row
#> 0 STD, City\tState
#> 1 33, Kolkata\tWest Bengal
#> 2 44, Chennai\tTamil Nadu
#> 3 40, Hyderabad\tTelengana
#> 4 80, Bangalore\tKarnataka

tmp = df["row"].str.split(",", n=1, expand=True)
df[["std", "city_state"]] = tmp.apply(lambda col: col.str.strip())
print(df)
```
